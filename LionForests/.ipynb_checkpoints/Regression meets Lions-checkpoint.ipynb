```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)

regr = RandomForestRegressor(max_depth=2, n_estimators=500, random_state=0)
regr.fit(X, y)

print(regr.predict([X[1]]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(num=None, figsize=(10, 6), dpi=100, facecolor='w', edgecolor='k')

# matplotlib histogram
plt.hist(y, color = 'blue', bins = int(50))
plt.hist(regr.predict(X), color = 'red', bins = int(50))

# seaborn histogram
sns.distplot(y, hist=True, kde=False, 
             bins=int(50), color = 'blue')
sns.distplot(regr.predict(X), hist=True, kde=False, 
             bins=int(50), color = 'red')
# Add labels
plt.title('Histogram of Y Values')
plt.xlabel('Instances')
plt.ylabel('Value')
```
